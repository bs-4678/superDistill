```python
# # -*- coding: utf-8 -*-
# # 请确保初始化了你的 client 变量
# # from 你的SDK import client
# import os
# from pathlib import Path
# from openai import OpenAI
# import time
# from datetime import datetime
# import argparse
# now = datetime.now()
# print(now)

# OPENAI_API_KEY = 'sk-ade5e567feba4013a46769eaeb4692b7'
# # 初始化客户端
# client = OpenAI(
#     api_key = OPENAI_API_KEY,
#     base_url="https://dashscope.aliyuncs.com/compatible-mode/v1"
# )

# batch_ids = [
#     "batch_7ad6f883-bcf2-4a8a-b03e-7468cd22918b",
#     "batch_2791d7c7-dc01-477e-ab3a-d8e5654d41fc",
#     "batch_f073c319-74fb-4459-bea7-5459f185fd40",
#     "batch_80b3a945-258e-4ac9-808a-83927dc8bb82",
#     "batch_de37fedc-cac5-4be6-a6c7-a080b59e6d0c",
#     "batch_f75a0793-e251-4b8c-885e-2d9fefea5abf",
#     "batch_636c17ee-1bad-49d7-9059-4b3e7d2a0792",
#     "batch_e4b9b383-3019-47ac-a470-db8aadec32b8",
#     "batch_71d372f8-7f2d-46cb-a292-864d33264d3c",
#     "batch_d2405323-e1f9-428d-91df-75cd23a5c11a",
#     "batch_bb83245d-cc53-42b4-955e-94d3c65e69b7",
#     "batch_8658da5e-8214-4e4c-a7fc-b87b87fa123f",
#     "batch_25e6bdac-10ad-42cb-8720-43700b0517ef",
#     "batch_78931b91-e146-4525-9229-c78caad04178",
#     "batch_fb21f811-044f-4d96-a930-a78e1ca0b4af",
#     "batch_61814064-1ed9-44d9-a1a9-3313c918de3e",
#     "batch_ac531ce1-8394-4684-ac8b-8dd97d999b5a",
#     "batch_39646f0b-d7e1-42ff-a937-31b4720ad004",
#     "batch_62b15ed7-224a-4702-931c-1e72638194c4",
#     "batch_43c58500-286e-4e3e-900c-64112fbdcb73"
# ]


# for batch_id in batch_ids:
#     try:
#         result = client.batches.cancel(batch_id)
#         print(f"{batch_id} 已成功取消")
#     except Exception as e:
#         print(f"{batch_id} 取消失败: {e}")






# DASHSCOPE_API_KEY1=sk-7bee41571d7b4ebfb5393d574dd1173f
# DASHSCOPE_API_KEY2=sk-ade5e567feba4013a46769eaeb4692b7
# DASHSCOPE_API_KEY3=sk-da18c7243e8d4fbb84aebb8567668139

import os
import requests




API_KEY = 'sk-ade5e567feba4013a46769eaeb4692b7'
BASE_URL="https://dashscope.aliyuncs.com/compatible-mode/v1"


def list_batches():
    url = f"{BASE_URL}/batches"
    headers = {"Authorization": f"Bearer {API_KEY}"}
    resp = requests.get(url, headers=headers)
    return resp.json().get("data", []) if resp.ok else []

def cancel_batch(batch_id):
    url = f"{BASE_URL}/batches/{batch_id}/cancel"
    headers = {"Authorization": f"Bearer {API_KEY}"}
    resp = requests.post(url, headers=headers)
    print(f"已取消 batch {batch_id}: 状态码 {resp.status_code}, 反馈 {resp.text}")

def main():
    batches = list_batches()
    print(len(batches), batches)
    if batches:
        for batch in batches:
            # 只取消状态不是 completed/failed/cancelled 的任务
            if batch.get("status") not in ["completed", "failed", "cancelled", "expired"]:
                cancel_batch(batch["id"])
    else:
        print("没有可操作的 batch 任务。")

if __name__ == "__main__":
    main()
```
